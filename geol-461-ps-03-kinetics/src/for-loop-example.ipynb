# Using for loops to create plots

## Import Python Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Using for loops

A for loop is useful when you want to create a plot that has curves for different values of a variable.  The example below uses the for loop to plot the equation:

$$ y = x^n $$

for the same range of x, but different values of n

Some tips when using for loops:
* replace the variable that you're looping through with `i` (in this case we replaced `n`)
* you can label your plot with `i` (and operations on i) to create a legend 
* call your plot labels and axis function after and outside of the for loop

```python
x = np.linspace(1, 10)
n = [2, 3, 4]

for i in n:
    y = x**i
    plt.plot(x, y, label = i)

plt.xlabel('X');
plt.ylabel('Y');
plt.yscale('log')
plt.legend(title = 'n=');
```
